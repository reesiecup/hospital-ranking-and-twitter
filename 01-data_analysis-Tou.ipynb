```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as sm
```

```python
data = pd.read_csv('./indi_handles_naming.csv', index_col = 0)
sorted_data = data.sort_values('Hospital Code')
```

```python
def calculate_pct(handle_chars_column):
    char = 0
    for i in range(len(sorted_data)):
        if sorted_data[handle_chars_column].iloc[i] == 'Yes':
            char += 1
    return char/len(sorted_data[handle_chars_column])

degree = calculate_pct('Handle_degree')
position = calculate_pct('Handle_position')
first_name = calculate_pct('Handle_first_name')
last_name = calculate_pct('Handle_last_name')
specialty = calculate_pct('Handle_specialty')

print('Twitter handle includes individual\'s degree: ', degree)
print('Twitter handle includes individual\'s position: ', position)
print('Twitter handle includes individual\'s first name: ', first_name)
print('Twitter handle includes individual\'s last name: ', last_name)
print('Twitter handle includes individual\'s specialty: ', specialty)
```

### Correlation between naming characteristics and the number of followers

```python
def reformat_chars(handle_chars_column, new_col_name):
    sorted_data[new_col_name] = 0
    sorted_data[new_col_name] = 0
    sorted_data[new_col_name][sorted_data[handle_chars_column] == 'Yes'] = 1
    return sorted_data

reformat_chars('Handle_degree', 'Handle_degree_bi')
reformat_chars('Handle_position', 'Handle_position_bi')
reformat_chars('Handle_first_name', 'Handle_first_name_bi')
reformat_chars('Handle_last_name', 'Handle_last_name_bi')
reformat_chars('Handle_specialty', 'Handle_specialty_bi')
```

```python
reg_data = sorted_data[['Followers', 'Handle_degree_bi', 'Handle_position_bi', 'Handle_first_name_bi', 
                        'Handle_last_name_bi', 'Handle_specialty_bi']]
result = sm.ols(formula="Followers ~ Handle_degree_bi + Handle_position_bi + Handle_first_name_bi + \
Handle_last_name_bi + Handle_specialty_bi", data = reg_data).fit()
print(result.summary())
```
