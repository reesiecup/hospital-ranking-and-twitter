### Requirements

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from scipy import stats
```

### Reading data

column __type_of_affiliation__: type of affiliation: 1 = private, 2 = public, 3 = mixed<br>

```python
data = pd.read_csv('./07-Quynh/official.csv')
type_of_affiliation = [1, 3, 1, 1, 2, 2, 2, 1, 1, 1, 3, 1, 1, 1, 2, 1, 1, 1, 1, 1]
affli_data = pd.Series(type_of_affiliation)
data['Affiliation'] = affli_data.values
data
```

### Check if two variables are normally distributed

```python
row_name =['Retweets to Likes', 'Affiliation']
for i in row_name:
    x = stats.shapiro(data[i])
    print('The test statistics and p-value for', i, 'is' , x)

# return (t-stat, p-value)
# Null: data is normal.
# If it's less than .05, not normal.

# Retweets to Likes is not normally distributed but affiliation types is normally distributed.
```

### One-way ANOVA

```python
private = [x for x in data['Retweets to Likes'][data['Affiliation'] == 1]]
public = [x for x in data['Retweets to Likes'][data['Affiliation'] == 2]]
mixed = [x for x in data['Retweets to Likes'][data['Affiliation'] == 3]]
stats.f_oneway(private, public, mixed) 
# No relation between affiliation types (private vs public vs mixed) and Twitter likability index
```

### Independent t-test

```python
public_mixed = public + mixed
stats.ttest_ind(private, public_mixed) 
# No relation between affiliation types (private vs public + mixed) and Twitter likability index
```
