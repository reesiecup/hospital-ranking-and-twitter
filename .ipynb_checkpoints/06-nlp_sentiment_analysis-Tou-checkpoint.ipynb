## Requirement

```python
import re
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Prepare the text data

```python
# Process tweets in institutional accounts
inst_list = [i for i in range(1, 21) if i != 7 and i != 11]
for i in inst_list:
    filename = './processed_data/inst_tweets_' + str(i) + '.csv'
    data = pd.read_csv(filename)
    data = data.drop(columns = ['Unnamed: 0']).rename(columns = {'0':'tweet'})
    data['processed_tweet'] = 'NaN'
    for j in range(len(data['tweet'])):
        raw_tweet = data['tweet'][j]
        in_processed_tweet = re.sub(r'@|#|(https?://[^\s]+)|(pic.twitter.com/[^\s]+)', '', raw_tweet)
        processed_tweet = re.sub('(\xa0[^\s+])|(\xa0)', '', in_processed_tweet)
        data['processed_tweet'][j] = processed_tweet
    out_filename = './sent_processed_data/inst_tweets_' + str(i) + '.csv'
    data.to_csv(out_filename)
```

```python
# Process tweets in individual accounts
indi_list = [i for i in range(1, 21) if i != 8]
for i in indi_list:
    filename = './processed_data/indi_tweets_' + str(i) + '.csv'
    data = pd.read_csv(filename)
    data = data.drop(columns = ['Unnamed: 0']).rename(columns = {'0':'tweet'})
    data['processed_tweet'] = 'NaN'
    for j in range(len(data['tweet'])):
        raw_tweet = data['tweet'][j]
        in_processed_tweet = re.sub(r'@|#|(https?://[^\s]+)|(pic.twitter.com/[^\s]+)', '', raw_tweet)
        processed_tweet = re.sub('(\xa0[^\s+])|(\xa0)', '', in_processed_tweet)
        data['processed_tweet'][j] = processed_tweet
    out_filename = './sent_processed_data/indi_tweets_' + str(i) + '.csv'
    data.to_csv(out_filename)
```

## Conduct preliminary analysis

```python
# Run sentiment analysis for institutional accounts
inst_list = [i for i in range(1, 21) if i != 7 and i != 11]
for i in inst_list:
    filename = './sent_processed_data/inst_tweets_' + str(i) + '.csv'
    data = pd.read_csv(filename, index_col = 0)
    data['polarity'] = 'NaN'
    data['subjectivity'] = 'NaN'
    for i in range(len(data['processed_tweet'])):
        tweet = TextBlob(data['processed_tweet'][i])
        data['polarity'][i] = tweet.sentiment.polarity
        data['subjectivity'][i] = tweet.sentiment.subjectivity
    data.to_csv(filename)
```

```python
# Run sentiment analysis for individual accounts
indi_list = [i for i in range(1, 21) if i != 8]
for i in indi_list:
    filename = './sent_processed_data/indi_tweets_' + str(i) + '.csv'
    data = pd.read_csv(filename, index_col = 0)
    data['polarity'] = 'NaN'
    data['subjectivity'] = 'NaN'
    for i in range(len(data['processed_tweet'])):
        tweet = TextBlob(data['processed_tweet'][i])
        data['polarity'][i] = tweet.sentiment.polarity
        data['subjectivity'][i] = tweet.sentiment.subjectivity
    data.to_csv(filename)
```

## Graph the frequency of each sentiment

```python
# Prepare institutional accounts data for frequency graphing
inst_list = [i for i in range(1, 21) if i != 7 and i != 11]
for i in inst_list:
    filename = './sent_processed_data/inst_tweets_' + str(i) + '.csv'
    data = pd.read_csv(filename, index_col = 0)
    data['Sentiment'] = 'NaN'
    data['Objectivity'] = 'NaN'
    for i in range(len(data['processed_tweet'])):
        if data['subjectivity'][i] > 0.5:
            data['Objectivity'][i] = 'Subjective'
        else:
            data['Objectivity'][i] = 'Objective'       
        if data['polarity'][i] > 0:
            data['Sentiment'][i] = 'Positive'
        elif data['polarity'][i] == 0:
            data['Sentiment'][i] = 'Neutral'
        else:
            data['Sentiment'][i] = 'Negative'
    data.to_csv(filename)
```

```python
# Prepare individual accounts data for frequency graphing
indi_list = [i for i in range(1, 21) if i != 8]
for i in indi_list:
    filename = './sent_processed_data/indi_tweets_' + str(i) + '.csv'
    data = pd.read_csv(filename, index_col = 0)
    data['Sentiment'] = 'NaN'
    data['Objectivity'] = 'NaN'
    for i in range(len(data['processed_tweet'])):
        if data['subjectivity'][i] > 0.5:
            data['Objectivity'][i] = 'Subjective'
        else:
            data['Objectivity'][i] = 'Objective'       
        if data['polarity'][i] > 0:
            data['Sentiment'][i] = 'Positive'
        elif data['polarity'][i] == 0:
            data['Sentiment'][i] = 'Neutral'
        else:
            data['Sentiment'][i] = 'Negative'
    data.to_csv(filename)
```
